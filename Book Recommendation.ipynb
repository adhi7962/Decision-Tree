```python
data= pd.read_csv(r'D:\Assignment\Recommendation System\book.csv')
```

```python
data_1=data.iloc[:,1:]
```

```python
data_2= data_1.rename(columns={'User.ID':'User','Book.Title':'Book','Book.Rating':'Rating'})
data_2
```

```python
# No of Unique user in dataset
len(data_2.User.unique())
```

```python
# No of Unique book in dataset
len(data_2.Book.unique())
```

```python
len(data_2.Rating.unique())
```

```python
user_book_df= pd.pivot_table(data_2,index='User',columns='Book',values='Rating')
user_book_df
```

```python
user_book_df.fillna(0,inplace=True)
```

```python
user_book_df
```

```python
#Calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user = 1-pairwise_distances(user_book_df.values,metric='cosine')
```

```python
user_sim_df= pd.DataFrame(user)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = data_2.User.unique()
user_sim_df.columns = data_2.User.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#IBCF
book_user_rating= user_book_df['10 Commandments Of Dating']
book_user_rating
```

```python
similar_to_book= user_book_df.corrwith(book_user_rating)
similar_to_book.head(10)
```

```python
corr_book = pd.DataFrame(similar_to_book, columns=['Correlation'])
corr_book.dropna(inplace=True)
corr_book.head(5)
```

```python
corr_book[corr_book['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
```
